```
from tensorflow.keras.models import load_model
# returns a compiled model
# identical to the previous one
model = load_model('/content/drive/My Drive/mri/2020-04-04_VGG_model.h5')
```

```
from keras.preprocessing.image import load_img
# load an image from file
image = load_img('/content/drive/My Drive/mri/Data_for_checking/699.jpg', target_size=(224, 224))
```

```
from keras.preprocessing.image import img_to_array
# convert the image pixels to a numpy array
image = img_to_array(image)
```

```
# reshape data for the model
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
```

```
from keras.applications.vgg16 import preprocess_input
# prepare the image for the VGG model
image = preprocess_input(image)
```

```
# predict the probability across all output classes
yhat = model.predict(image)
```

```
import numpy as np

classes = ["glioma", "pituitary", "meningioma", "Norm"]

print(yhat)
print(classes[np.argmax(yhat)])
```
